```python
# Beta => Covariance / Variance    OR    (Correlation * Stock_STD) / StandardMarket_STD

stock_beta = pd.Series(index=stock_list)
nifty_std = corr_stock['nifty_50'].std()

for stock in stock_list:
    stock_beta[stock] = (corr_df['nifty_50'][stock] * corr_stock[stock].std()) / nifty_std
```

```python
metrics = ['Return','Volatility','Sharpe Ratio', 'VaR', 'Portfolio Beta', 'Treynor Ratio', 'Jensen Alpha']
p = ['Portfolio 1', 'Portfolio 2', 'Portfolio 3', 'Portfolio 4', 'Portfolio 5', 'Portfolio 6','Portfolio 7', 'Portfolio 8', 'Portfolio 9', 'Portfolio 10']
portfolio = []

max_sharpe = pd.DataFrame(index=metrics)
```

```python
while len(portfolio) < 10:
    sm = stock_list.copy()
    sl = sample(stock_list,5)
    sl.sort()
    if sl in portfolio:
        continue
    else:
        portfolio.append(sl)
    for st in sl:
        sm.remove(st)
    portfolio.append(sm)
```

```python
portfolio
```

```python
for portfolios in portfolio:

    daily_retn = daily_ret[portfolios].copy()
    cov_matrixn = daily_retn.cov()
    stock_betan = stock_beta[portfolios].copy()
    
    
    runs = 10
    results = np.zeros((7+len(portfolios),runs))
    risk_free_rate = 0.0602
    alpha = 0.01
    days = 252
#     total_investment = 1000000
    market_return = nifty_50_daily.mean() * days 

    for i in range(runs):

        pb = 0

        weights = np.random.random(len(portfolios))
        weights /= np.sum(weights)

        portfolio_return = np.sum(daily_retn.mean()*weights) * days
        portfolio_std_dev = np.sqrt(np.dot(weights.T,np.dot(cov_matrixn, weights))) * np.sqrt(days)

        results[0,i] = portfolio_return
        results[1,i] = portfolio_std_dev

        # SHARPE RATIO => (portfolio_return - risk_free_rate) / volatility
        results[2,i] = (results[0,i] - risk_free_rate) / results[1,i]

        # VaR => return - (Z * volatility)
        results[3,i] = abs(portfolio_return - (portfolio_std_dev * norm.ppf(1 - alpha)))  # ppf => inverse of cdf

        for j in range(len(weights)):
            results[j+7,i] = weights[j]
            pb += weights[j] * stock_betan[j]

        # PORTFOLIO BETA => sum(weight * beta)
        results[4,i] = pb

        # TREYNOR RATIO => (portfolio_return - risk_free_rate) / portfolio_beta
        results[5,i] = (results[0,i] - risk_free_rate) / results[4,i]

        # JENSEN ALPHA => portfolio_return - (risk_free_rate + portfolio_beta * (market_return - risk_free_rate))
        results[6,i] = results[0,i] - (risk_free_rate + results[4,i] * (market_return - risk_free_rate))
        
    
    a = metrics+portfolios
    results = pd.DataFrame(results.T, columns=a)
    
    total_results = results.copy()
    for stock in portfolios:
        total_results[stock] = total_results[stock] * 100
     
    
    total_results = pd.concat([total_results[metrics], total_results[portfolios].round(decimals=2)], axis=1)
    
    max_sharpen = total_results.iloc[results['Sharpe Ratio'].idxmax()]
    
    max_sharpe = pd.concat([max_sharpe, max_sharpen], axis=1)
    
    print(max_sharpen.to_frame().T)
max_sharpe.columns = p
```
